### Filtering with 'And (&)' operator

```python
import pandas as pd
empl = pd.read_csv('1.1 pandas.zip/employees.csv', parse_dates =['Start Date', 'Last Login Time']) #convert date and time to pandas readables
empl['Gender'] = empl['Gender'].astype('category')
empl['Senior Management'].astype('bool')
empl
```

```python
#Getting values with 2 or more conditions using (And) operator
gen = empl['Gender'] == 'Male' 
tim = empl['Team'] == 'Sales'

empl[gen & tim].head(5)
```

### Filtering with 'OR ( | )' operator

```python
#Find all Senior Mangers or those who stared before 1990

sm = empl['Senior Management'] == True
sd = empl['Start Date'] < '1991-01-01'
```

```python
empl[sm | sd]
```

```python
#Get the name of 'Robert'  and  'Team: Client Services' or start > jun 1, 2000

nm = empl['First Name'] == 'Robert'
tim = empl['Team'] == 'Client Services'
date = empl['Start Date'] > '2016-06-01'

empl[nm & (tim | date)]  #the perenthesis make it run first
```

### Filtering with '.isin( )'

```python
#get inof on only these items ('Marketing', 'Legal', 'Finance')?
empl[fav].head(3)
```

### Filtering with 'between( ) method'
- this used to find the list of items within a range

```python
#Find employees with a salary of between 60000 to 900000

newSal = empl['Salary'].between(60000, 70000)

empl[newSal].head(5)
```

```python
#Find employees with a bonus of between 2 and 5 %
bonus = empl['Bonus %'].between(2.0,5.0)
empl[bonus]
```

```python
login = empl['Last Login Time'].between('8:00 AM','12:00 PM')
empl[login].head(10)
```

### Filtering with 'duplicate( ) method'
- this used to find the duplicates in a Series
- i.e. it finds items that occur more than once. eg: it 'A' appears 4 times, the first instance is kept and the other 3 are considered as duplicats

```python
#Find duplcates in Name
empl.sort_values('First Name', inplace=True)
empl['First Name']

empl['First Name'].duplicated() #idnetifies name that appears more than once. Use keep = 'first' / 'Last' to determine which duplucate u want to keep
```

```python
# Remove duplicates
empl['First Name'].duplicated(keep = False) #this keeps all the duplicates

#use the tilda (~) sing to reverse the Series to get only items that appear only once

single = ~empl['First Name'].duplicated(keep = False)
```

```python
empl[single]
```

### Filtering with 'drop_duplicate( ) method'
- this used to remove duplicates from a Series

```python
empl.drop_duplicates(subset=['First Name'], keep = False) # False removes any instance of the found duplicates
# the drop_duplicates() method can also search through multiple columns to find duplicates

empl.drop_duplicates(subset=['First Name', 'Team'], keep = 'first')
```
