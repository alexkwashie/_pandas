 ### The astype( ) Method
 - the astype() doesnt work with NaN values, so all NaN need to be removed

```python
import pandas as pd
nba1 = pd.read_csv("1.1 pandas.zip/nba.csv").dropna(how = 'all')
#dropna(how='all') is used to remove a row with all NaN in it as seen on the last row
nba1["Salary"].fillna(0, inplace= True)
nba1["College"].fillna('No college', inplace = True)
```

```python
nba1.tail(5)
```

```python
#pandas automatically converts intergers to float numbers, so use the astype() to convet it bact to integers
nba1["Salary"].astype("int")

# the astype() doesnt have an inplace arg, so will have reasigne it the main variable to store the change
```

```python
nba1["Salary"] = nba1["Salary"].astype("int")
```

```python
nba1["Age"] = nba1["Age"].astype("int")
```

```python
nba1["Number"] = nba1["Number"].astype("int")
```

```python
nba1.dtypes
nba1.info()  #this gives u count of all rows and the type of data it is 
```

#### Using categories
- this i used to reduce the memeory used by the data provided. it does this by putting it into categories and usethe copy of a that word in each row. eg. male or femail/ days of the weeks so it doesnt have to store each string on each row.

```python
# find the categories in a column with the nunique() function 
nba1 ["Position"].nunique()
```

```python
#use the category on the nba1["Position"]
nba1["Position"] = nba1["Position"].astype("category")
```

### Sort a dataframe with .sort_value 1

```python
import pandas as pd
nba2 = pd.read_csv("1.1 pandas.zip/nba.csv")
nba2
```

```python
nba2.sort_values("Name") #ascending=False means reversing it i.e from bottom to top
#sorting by column
```

```python
nba2.sort_values("College", na_position='last') #first
#na_position is use to place the NaN values at the top or bottom of the sorted list
```

### Sort a dataframe with .sort_value 2

```python
# Sorting multiple rows
nba2.sort_values(["Team", "Name"], ascending=(True, False))
# ascending reversing the respective column
```

### Sort a dataframe with .sort_index

```python
nba2.sort_index(ascending= False)
```
