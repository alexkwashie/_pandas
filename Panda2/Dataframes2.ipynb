### Memory Management etc.

```python
import pandas as pd
df = pd.read_csv('1.1 pandas.zip/employees.csv')

#RATHER THAN USING (pd.to_datetime) to convert to date time like as seen below, u can do it in the (pd.read_csv)
# df = pd.read_csv('1.1 pandas.zip/employees.csv', parse_dates = ['Start Date',Last Login Time'])
```

```python
df.head()
```

```python
# Convert date to a date-time object
df['Start Date'] = pd.to_datetime(df['Start Date'])
```

```python
# Convert Last login Time to a date-time object
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
```

```python
# Convert Gender to a catergory field to save memeory
df['Gender'] = df['Gender'].astype('category')
```

```python
# Convert Senior Management to a boolean field to save memeory
df['Gender'] = df['Gender'].astype('bool')
```

### Filtering dataframe base on condition

```python
import pandas as pd
df1 = pd.read_csv('1.1 pandas.zip/employees.csv')
```

```python
# Find all Male values
df1["Gender"] == 'Male'
```

```python
#Extract information on all Male values
df1[df1["Gender"] == 'Male']
```

```python
#Extract information on all in Marketing department
mdept = df1["Team"] == 'Marketing'
```

```python
df1[mdept]
```

### Filtering with 'And (&)' operator

```python
import pandas as pd
empl = pd.read_csv('1.1 pandas.zip/employees.csv', parse_dates =['Start Date', 'Last Login Time']) #convert date and time to pandas readables
empl['Gender'] = empl['Gender'].astype('category')
empl['Senior Management'].astype('bool')
empl
```

```python
#Getting values with 2 or more conditions using (And) operator
gen = empl['Gender'] == 'Male' 
tim = empl['Team'] == 'Sales'

empl[gen & tim]
```
