```python
# Convert date to a date-time object
df['Start Date'] = pd.to_datetime(df['Start Date'])
```

```python
# Convert Last login Time to a date-time object
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
```

```python
# Convert Gender to a catergory field to save memeory
df['Gender'] = df['Gender'].astype('category')
```

```python
# Convert Senior Management to a boolean field to save memeory
df['Gender'] = df['Gender'].astype('bool')
```
